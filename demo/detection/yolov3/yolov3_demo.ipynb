```
# import the yolo model 

from PyVision.pyvision.detection import yolov3

import matplotlib.pyplot as plt
```

```
# We are going to use images that come with the PyVision 
# repo but feel free to use your own

examples_path = "PyVision/tests/detection/yolov3/"
```

```
# let us use the default yolov3-416 model. For more information, refer to the docs

yolo = yolov3.YOLOv3(device="gpu", save=False, show=False)
```

```
# we run inference with the detect method
time_taken, det_img, det_objs = yolo.detect(examples_path+"zebra_test.jpg")

# Printing the detected objects dictionary for demonstration
print("Time Taken: {}\nLen of object dict: {}\n{}".format(time_taken, len(det_objs), det_objs))

# showing the original image for reference
plt.imshow(plt.imread(examples_path+"zebra_test.jpg"))
plt.show()

# showing the image with bounding boxes drawn on them automatically since 
# parameter draw=True
plt.imshow(det_img)
plt.show()
```

```
# we run inference with the detect method
time_taken, det_img, det_objs = yolo.detect(examples_path+"cars_test.jpg")

# Printing the detected objects dictionary for demonstration
print("Time Taken: {}\nLen of object dict: {}\n{}".format(time_taken, len(det_objs), det_objs))

# showing the original image for reference
plt.imshow(plt.imread(examples_path+"cars_test.jpg"))
plt.show()

# showing the image with bounding boxes drawn on them automatically since 
# parameter draw=True
plt.imshow(det_img)
plt.show()
```

```
# let us use the yolov3-tiny model. For more information, refer to the docs

yolo_tiny = yolov3.YOLOv3(model="yolov3-tiny", device="gpu", save=False, show=False)
```

```
# we run inference with the detect method
time_taken, det_img, det_objs = yolo_tiny.detect(examples_path+"zebra_test.jpg")

# Printing the detected objects dictionary for demonstration
print("Time Taken: {}\nLen of object dict: {}\n{}".format(time_taken, len(det_objs), det_objs))

# showing the original image for reference
plt.imshow(plt.imread(examples_path+"zebra_test.jpg"))
plt.show()

# showing the image with bounding boxes drawn on them automatically since 
# parameter draw=True
plt.imshow(det_img)
plt.show()
```
