# Neural Style Transfer  

This standalone colab notebook demonstrates how you can use the Neural Style Transfer Model in just 3 Lines! 

While this notebook uses the images that come by default in the PyVision repo, feel free to use your own images.

```
# We import Neural Style Transfer Module

from PyVision.pyvision.misc import NeuralStyleTransfer

# For displaying the images here 
import matplotlib.pyplot as plt
```

```
# While for purposes of demonstration we use images that come with
# the repo, feel free to use your own images! 

example_path = "PyVision/pyvision/misc/NeuralStyleTransfer/Examples/images/"
```

```
# We gotta initialize the Neural style transfer model first

nst = NeuralStyleTransfer.NeuralStyle(use_gpu=True)
```

```
# Let us store the paths of the style image and the content image 
# Note: While this notebook uses the images that come by default with 
# the PyVision Repo, feel free to use your own images. 

content_img = example_path+"content3.jpg"
style_img = example_path+"style7.jpg"

stylized_img, time_taken = nst.run_style_transfer(style_img=style_img, content_img=content_img)

print("Time Taken: ", time_taken)

# Displaying the original content image for reference 
plt.imshow(plt.imread(content_img))
plt.show()

# Displaying the original style image for reference
plt.imshow(plt.imread(style_img))
plt.show()

# Now, displaying the stylized image
plt.imshow(stylized_img)
plt.show()
```

```
# But, you can also tinker with the various parameters. For example, 
# you can play around with "style_weight" and "content_weight" hyperparameters
# to give the desired relative importance to "style" and "content"
# You can also play around with WHICH layers in the network you want to use 
# as the style and content layers by using the "content_layers" and "style_layers"
# parameters

nst2 = NeuralStyleTransfer.NeuralStyle(use_gpu=True, content_layers=['conv_5'], style_weight=1e6, content_weight=2)
```

```
# Let us store the paths of the style image and the content image 
# Note: While this notebook uses the images that come by default with 
# the PyVision Repo, feel free to use your own images. 

content_img = example_path+"content3.jpg"
style_img = example_path+"style7.jpg"

stylized_img, time_taken = nst2.run_style_transfer(style_img=style_img, content_img=content_img)

print("Time Taken: ", time_taken)

# Displaying the original content image for reference 
plt.imshow(plt.imread(content_img))
plt.show()

# Displaying the original style image for reference
plt.imshow(plt.imread(style_img))
plt.show()

# Now, displaying the stylized image
plt.imshow(stylized_img)
plt.show()
```
